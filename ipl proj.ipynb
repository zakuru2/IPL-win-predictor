```python
delivery["cumulative_sum"]=delivery.groupby(["match_id","inning"]).cumsum(numeric_only='True')["total_runs"]#this is done as it will be used later to calculate runs left
```

```python
delivery.head()
```

```python
teams=[
    "Sunrisers Hyderabad","Mumbai Indians",
    "Royal Challengers Bangalore","Kolkata Knight Riders",
    "Kings XI Punjab","Chennai Super Kings","Rajasthan Royals",
    "Delhi Capitals"
] # as many teams are not playing currently
```

```python
match["team2"]=match["team1"].str.replace("Delhi Daredevils","Delhi Capitals")
match["team1"]=match["team1"].str.replace("Delhi Daredevils","Delhi Capitals")
match["team2"]=match["team1"].str.replace("Deccan Chargers","Sunrisers Hyderabad")
match["team1"]=match["team1"].str.replace("Deccan Chargers","Sunrisers Hyderabad")
new_match=new_match[new_match["team1"].isin(teams)]
new_match=new_match[new_match["team2"].isin(teams)]
```

```python
new_match.head()
```

```python
new_match["team1"].unique()
```

```python
delivery.head()
```

```python
new_match=new_match.drop(["team1","team2","venue"],axis=1)
```

```python
new_match.head()
```

```python
delivery=delivery.merge(new_match,on="match_id")
```

```python
delivery=delivery[delivery["inning"]==2]
```

```python
delivery
```

```python
delivery["runs_left"]=delivery["total_runs_y"]-delivery["cumulative_sum"]
```

```python
delivery
```

```python
delivery["balls_left"]=126-(delivery["over"]*6 + delivery["ball"])
```

```python
delivery
```

```python
delivery["player_dismissed"]=delivery["player_dismissed"].fillna("0")
delivery["player_dismissed"]=delivery["player_dismissed"].apply(lambda x:x if x=="0" else "1")
delivery["player_dismissed"]=delivery["player_dismissed"].astype(int)
delivery["player_dismissed"]=delivery.groupby(["match_id"]).cumsum(numeric_only="True")["player_dismissed"]
delivery["wickets_rem"]=10-delivery["player_dismissed"]
```

```python
delivery
```

```python
delivery["crr"]=(delivery["cumulative_sum"]*6)/(120-delivery["balls_left"])
```

```python
delivery
```

```python
delivery.head(10)
```

```python
delivery["rrr"]=(delivery["runs_left"]*6)/delivery["balls_left"]
```

```python
delivery
```

```python
def convert(obj):
    if obj["batting_team"]==obj["winner"]:
        return 1
    else:
        return 0
    
```

```python
delivery["winner"]=delivery.apply(convert,axis=1)
```

```python
delivery
```

```python
delivery_new=delivery[["batting_team","bowling_team","city","winner","total_runs_y","runs_left","balls_left","wickets_rem","crr","rrr"]]
```

```python
delivery_new
```

```python
delivery_new["total_runs"]=delivery_new["total_runs_y"] #this is giving error as  Pandas is warning you that you might be modifying a copy rather than the original DataFrame. To avoid this warning, you should use the .loc
```

```python
delivery_new=delivery_new[delivery_new["runs_left"]>=0]
```

```python
delivery_new=delivery_new.sample(delivery_new.shape[0])#this shuffles the whole dataframe
```

```python
delivery_new.head()
```

```python
delivery_new["result"]=delivery_new["winner"]
```

```python
delivery_new=delivery_new.drop(["winner","total_runs_y"],axis=1)
```

```python
delivery_new.head()
```

```python
delivery_new.isnull().sum()
```

```python
delivery_new.dropna(inplace=True)
```

```python
delivery_new = delivery_new.replace([np.inf, -np.inf], np.nan)  # Replacing inf with NaN in column 'A'
delivery_new.dropna(inplace=True)
```

```python
x=delivery_new.iloc[:,:-1]
y=delivery_new.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import make_pipeline
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
```

```python
 x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
x_train.head()
```

```python
trf=ColumnTransformer([
    ("trf",OneHotEncoder(sparse=False,drop="first"),["batting_team","bowling_team","city"])
],remainder="passthrough") #A sparse matrix is a matrix in which most of the elements are zero
```

```python
lo=LogisticRegression(solver="liblinear")
```

```python
pipe=make_pipeline(trf,lo)
```

```python
y_train
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred=pipe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_pred,y_test))
```
